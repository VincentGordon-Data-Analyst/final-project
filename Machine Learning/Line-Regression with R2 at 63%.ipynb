```python
import numpy as np
import pandas as pd 
from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("../Resources/housing.csv")
```

```python
df
```

```python
df=df.ffill()
```

```python
df.head()
```

```python
df_y=df.drop(columns='median_house_value',axis=1)
df_y
```

```python
df_encoded = pd.get_dummies(df_y, columns=['ocean_proximity'])
df_encoded.head()
```

```python
df_data_scaled = StandardScaler().fit_transform(df_encoded)
```

```python
df_data_scaled
```

```python
df_encoded.columns
```

```python
df_data_scaled_transformed=pd.DataFrame(df_data_scaled,columns=df_encoded.columns)
df_data_scaled_transformed
```

```python
# End of Data cleaning here
```

```python
X=df_data_scaled_transformed
y=df['median_house_value']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1)
```

```python
model1 = LinearRegression()
```

```python
model1.fit(X_train,y_train)
```

```python
LinearRegression() 
```

```python
y_predict=model1.predict(X_test)
```

```python
score = model1.score(X_test, y_test,sample_weight=None)
```

```python
score
```

```python
r2 = r2_score(y_test, y_predict)
```
